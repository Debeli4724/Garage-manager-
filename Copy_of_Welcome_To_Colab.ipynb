```

# ==========================================
# WORLDQUANT UNIVERSITY STYLE PROJECT
# Portfolio Analysis Program
# ==========================================

# Import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# ------------------------------------------
# STEP 1: Create Stock Data
# ------------------------------------------

data = {
    "Month": ["Jan", "Feb", "Mar", "Apr", "May", "Jun"],
    "Apple": [150, 155, 160, 158, 165, 170],
    "Tesla": [200, 210, 220, 215, 225, 230],
    "Amazon": [120, 125, 130, 128, 135, 140]
}

# Convert to DataFrame
df = pd.DataFrame(data)

# Set Month as index
df.set_index("Month", inplace=True)

print("STOCK PRICE DATA")
print(df)

# ------------------------------------------
# STEP 2: Calculate Monthly Returns
# ------------------------------------------

returns = df.pct_change()

print("\nMONTHLY RETURNS")
print(returns)

# ------------------------------------------
# STEP 3: Portfolio Weights
# ------------------------------------------

weights = np.array([0.4, 0.3, 0.3])

print("\nPORTFOLIO WEIGHTS")
print("Apple :", weights[0])
print("Tesla :", weights[1])
print("Amazon:", weights[2])

# ------------------------------------------
# STEP 4: Expected Portfolio Return
# ------------------------------------------

mean_returns = returns.mean()

portfolio_return = np.sum(mean_returns * weights)

print("\nAVERAGE RETURNS")
print(mean_returns)

print("\nEXPECTED PORTFOLIO RETURN")
print(portfolio_return)

# ------------------------------------------
# STEP 5: Portfolio Risk
# ------------------------------------------

cov_matrix = returns.cov()

portfolio_variance = np.dot(weights.T,
                            np.dot(cov_matrix, weights))

portfolio_risk = np.sqrt(portfolio_variance)

print("\nPORTFOLIO RISK")
print(portfolio_risk)

# ------------------------------------------
# STEP 6: Plot Stock Prices
# ------------------------------------------

df.plot(figsize=(10,6))

plt.title("Stock Prices Over Time")
plt.xlabel("Month")
plt.ylabel("Price")
plt.grid(True)

plt.show()

# ------------------------------------------
# STEP 7: Plot Portfolio Returns
# ------------------------------------------

portfolio_returns = returns.dot(weights)

portfolio_returns.plot(
    kind='bar',
    figsize=(8,5)
)

plt.title("Portfolio Monthly Returns")
plt.xlabel("Month")
plt.ylabel("Return")

plt.grid(True)

plt.show()

# ------------------------------------------
# STEP 8: Final Investment Simulation
# ------------------------------------------

initial_investment = 10000

final_value = initial_investment * (1 + portfolio_return)

print("\nINITIAL INVESTMENT =", initial_investment)
print("ESTIMATED FINAL VALUE =", final_value)

# ------------------------------------------
# END PROGRAM
# ------------------------------------------
```
