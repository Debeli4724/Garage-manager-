```python

# =====================================================
# SIMPLE PORTFOLIO ANALYSIS PROJECT
# QUANTITATIVE FINANCE PRACTICE
# =====================================================

# -----------------------------------------------------
# STEP 1 : IMPORT LIBRARIES
# -----------------------------------------------------

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# -----------------------------------------------------
# STEP 2 : CREATE MARKET DATA
# -----------------------------------------------------

market_data = {

    "Month": ["Jan", "Feb", "Mar", "Apr", "May", "Jun"],

    "Google": [100, 105, 110, 108, 115, 120],

    "Microsoft": [90, 92, 95, 97, 100, 105],

    "Netflix": [80, 82, 85, 84, 88, 92]
}

# Create DataFrame
stocks = pd.DataFrame(market_data)

# Set Month as index
stocks.set_index("Month", inplace=True)

# Display stock prices
print("=" * 60)
print("MARKET STOCK DATA")
print("=" * 60)

print(stocks)

# -----------------------------------------------------
# STEP 3 : CALCULATE RETURNS
# -----------------------------------------------------

returns = stocks.pct_change()

# Remove NaN values
returns = returns.dropna()

print("\n" + "=" * 60)
print("MONTHLY STOCK RETURNS")
print("=" * 60)

print(returns)

# -----------------------------------------------------
# STEP 4 : DEFINE INVESTMENT WEIGHTS
# -----------------------------------------------------

portfolio_weights = np.array([0.5, 0.3, 0.2])

print("\n" + "=" * 60)
print("PORTFOLIO WEIGHTS")
print("=" * 60)

print("Google     :", portfolio_weights[0])
print("Microsoft  :", portfolio_weights[1])
print("Netflix    :", portfolio_weights[2])

# -----------------------------------------------------
# STEP 5 : CALCULATE AVERAGE RETURNS
# -----------------------------------------------------

average_returns = returns.mean()

print("\n" + "=" * 60)
print("AVERAGE RETURNS")
print("=" * 60)

print(average_returns)

# -----------------------------------------------------
# STEP 6 : EXPECTED PORTFOLIO RETURN
# -----------------------------------------------------

expected_return = np.sum(
    average_returns * portfolio_weights
)

print("\n" + "=" * 60)
print("EXPECTED PORTFOLIO RETURN")
print("=" * 60)

print(round(expected_return, 4))

# -----------------------------------------------------
# STEP 7 : CALCULATE PORTFOLIO RISK
# -----------------------------------------------------

covariance_matrix = returns.cov()

portfolio_variance = np.dot(
    portfolio_weights.T,
    np.dot(covariance_matrix,
           portfolio_weights)
)

portfolio_risk = np.sqrt(portfolio_variance)

print("\n" + "=" * 60)
print("PORTFOLIO RISK")
print("=" * 60)

print(round(portfolio_risk, 4))

# -----------------------------------------------------
# STEP 8 : STOCK PRICE VISUALIZATION
# -----------------------------------------------------

stocks.plot(figsize=(10, 6))

plt.title("Stock Price Movement")
plt.xlabel("Month")
plt.ylabel("Price")
plt.grid(True)

plt.show()

# -----------------------------------------------------
# STEP 9 : PORTFOLIO RETURNS
# -----------------------------------------------------

portfolio_monthly_returns = returns.dot(
    portfolio_weights
)

print("\n" + "=" * 60)
print("PORTFOLIO MONTHLY RETURNS")
print("=" * 60)

print(portfolio_monthly_returns)

# -----------------------------------------------------
# STEP 10 : BAR CHART OF RETURNS
# -----------------------------------------------------

portfolio_monthly_returns.plot(
    kind='bar',
    figsize=(8, 5)
)

plt.title("Portfolio Return Per Month")
plt.xlabel("Month")
plt.ylabel("Return")
plt.grid(True)

plt.show()

# -----------------------------------------------------
# STEP 11 : INVESTMENT GROWTH SIMULATION
# -----------------------------------------------------

starting_money = 5000

future_money = starting_money * (
    1 + expected_return
)

print("\n" + "=" * 60)
print("INVESTMENT RESULTS")
print("=" * 60)

print("Starting Investment :", starting_money)

print("Estimated Future Value :",
      round(future_money, 2))

# -----------------------------------------------------
# END OF PROJECT
# -----------------------------------------------------
```
